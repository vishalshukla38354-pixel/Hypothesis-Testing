Question 1: What is a null hypothesis (H₀) and why is it important in hypothesis testing?

 Ans:

  
  The  **null hypothesis(H₀)** is a statement that assumes there is **no effect , no difference , or no  relationship** between variables .

  It is  important because it provides a starting point for statistical testing. Researchers test whether there is enough evidence to reject H₀.

  Example":
  -

  H₀: The average marks of students are 70.

Question 2: What does the significance level (α) represent in hypothesis testing?

Ans:

The s**ignificance lavel(α)** is the probability of rejecting the null hypothesis when it is actually true.

Common values are:

- 0.05(5%)
- 0.01(1%)

if the p- value is less than α, we reject H₀.


Question 3: Differentiate between Type I and Type II errors


Error Type	  and   Meaning
-

Type I Error ➡	Rejecting a true null hypothesis

Type II Error ➡	Failing to reject a false null hypothesis

Simple Example
-
- Type I Error: Saying a person is guilty when they are innocent.

- TypeII Error: Saying a guilty person is innocent.


Question 4: Explain the difference between a one-tailed and two-tailed test. Give an example of each.

Ans:

One-tailed Test
-
Used when the hypothesis checks only one direction (greather then or less then).

Example:
-
H₁: Average salary > ₹50,000

Two-tailed Test
-
Used when the hypothesis checks both directions (not equal to).

Example:
-

Question 5: A company claims that the average time to resolve a customer complaint is 10 minutes.
A random sample of 9 complaints gives an average time of 12 minutes and a standard deviation of 3 minutes.
At α = 0.05, test the claim.

- Given:

- Claimed mean = 10 minutes
- sample mean = 12 minutes
- Sample standard deviation = 3 minutes
- Sample size = 9
- α = 0.05

since sample size is small (n<30), use a **t-test**.

**Hypotheses**

- H₀: μ = 10

- H₁: μ ≠ 10

**Test Statistic**

t = xˉ - μ / s/√n

Substitute values:

t = 12-10/3/√9

t = 2/1 = 2

Degrees of freedom:

df = n-1 = 8

Critical value at a = 0.05(two-tailed) = ±2.306

Since:

 2<2.306

Conclusion
-
Fail to reject H₀.


There is not enough evidence to say that the average complaint resolution time is different from 10 minutes.

Question 6: When should you use a Z-test instead of a t-test?

- Use a **Z-test** when:
1. The sample size is large (N≥30), ore
2. The population standard deviation is known.

Use a **t-test** when:

1. The sample size is small(n<30), and
2. The population standard deviation is unknown.

Question 7: The productivity of 6 employees was measured before and after a training program. At α = 0.05, test if the training improved productivity

**Step2:** **Find Differences (After - Before)**

 ∑d = 18

 sd ​≈ 1.79

 Test statistic:

 t =  dˉ/ sd/√n

t = 3/1.79/√6

t ≈ 4.10

Degrees of freedome:

df = n - 1 = 5

Reject H₀.

There is sufficient evidence that the training improved productivity.
